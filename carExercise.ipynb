```python
df= pd.read_csv('car_data.csv')
df.head()
```

### 1. Find the best selling Audi car in Europe

```python
audi = df[(df['Origin'].isin(['Europe'])) & (df['Make'].isin(['Audi']))]
audi
```

```python
# Total cars counts:
a = audi['Model'].value_counts()
print("Total no.of audi cars:" , len(a))
print("\n")
print(a)
```

```python
# find audi count type
import re
txt = audi['Model']
a = []
c = []
for i in txt:
    b = re.compile(r'A4')
    val = b.search(i)
    a.append(val)
```

```python
# Remove the none values:
car = []
no = []
for i in a:
    if i == None:
        no.append(i)
    else:
        car.append(i)
try:
    del no
    print(no)
except NameError:
    print("Succefully deleted none values")
```

```python
print("Totla no.of Audi A4 cars: ",len(car))
```

###  Answer :Audi a4 model car is the best sold car in Europe
